### **Copying dataset**(not to lose original data) **and Importing sklearn modules:**

```python
filtered_dataset = df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

### **Ordinal Encoding for Education column:**

```python
filtered_dataset['education'].unique()
```

```python
priority = [
    'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th',
    '11th', '12th', 'HS-grad', 'Some-college', 'Assoc-voc',
    'Assoc-acdm', 'Bachelors', 'Masters', 'Prof-school', 'Doctorate'
]
```

```python
# Now doing Ordinal Encoding on 'Education' column
oe = OrdinalEncoder(categories=[priority])
oe.fit(filtered_dataset[['education']])
```

```python
filtered_dataset['education'] = oe.transform(filtered_dataset[['education']]).astype(np.int32)
```

```python
filtered_dataset.head()
```

```python
filtered_dataset.drop(columns=['educational-num'],inplace=True) # since they both represent same, removing
# either of it is better
```

### **Label Encoding for Income column:**

```python
filtered_dataset['income'].unique()
```

```python
le = LabelEncoder()
filtered_dataset['income'] = le.fit_transform(filtered_dataset['income'])
```

```python
filtered_dataset.head() # <=50K mean 0 and >50K mean 1
```

### **One Hot Encoder for Non-ranked Categorical Columns:**

```python
filtered_dataset.head()
```

```python
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False,drop='first')
```

```python
filtered_dataset.shape
```

```python
categorical_cols = [
    'workclass',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'gender',
    'native-country'
]
```

```python
encoded = ohe.fit_transform(filtered_dataset[categorical_cols])

encoded_df = pd.DataFrame(
    encoded,
    columns=ohe.get_feature_names_out(categorical_cols)
)
encoded_df.head()
```

```python
original_df_dropped = filtered_dataset.copy()
original_df_dropped.drop(columns=categorical_cols,inplace=True)
original_df_dropped.head()
```

```python
final_dataset_after_ohe = pd.concat(
    [original_df_dropped, encoded_df],
    axis=1
)
```

```python
final_dataset_after_ohe.head()
```

```python
final_dataset_after_ohe.shape
```

### **Scaling Numerical Features using StandardScaler:**

```python
numerical_cols = [
    'age',
    'fnlwgt',
    'education',
    'capital-gain',
    'capital-loss',
    'hours-per-week'
]
```

```python
scaler = StandardScaler()
```

```python
scaled_output = scaler.fit_transform(final_dataset_after_ohe[numerical_cols])
```

```python
scaled_output
```

```python
scaled_df = pd.DataFrame(
    scaled_output,
    columns=numerical_cols,
    index=final_dataset_after_ohe.index
)
```

```python
scaled_df
```

```python
final_dataset_after_scaling = final_dataset_after_ohe.copy()
```

```python
final_dataset_after_scaling[numerical_cols] = scaled_df
```

```python
final_dataset_after_scaling.head()
```

#### **Model Readiness: Before vs After Scaling**

---

##### **Before Scaling**

###### **Observations**
- Numerical features have **very different ranges**
- `fnlwgt` and `capital-gain` have **very large values**
- `age` and `hours-per-week` have **small values**

###### **Problems**
- Large-value features **dominate the model**
- Distance-based models behave poorly
- Gradient-based models converge slowly

###### **Model Readiness**
❌ Not model-ready

---

##### **After Scaling (StandardScaler)**

###### **Observations**
- All numerical features have:
  - Mean ≈ 0
  - Standard deviation ≈ 1
- Feature values are on a **similar scale**

###### **Improvements**
- No feature dominates due to size
- Distance calculations become fair
- Faster and stable model training

###### **Model Readiness**
✅ Model-ready

---

##### **Comparison Summary**

| Aspect | Before Scaling | After Scaling |
|------|---------------|---------------|
| Feature scale | Uneven | Standardized |
| Feature dominance | Present | Removed |
| Distance-based models | Poor | Good |
| Gradient-based models | Slow | Stable |
| Overall readiness | ❌ No | ✅ Yes |

---

##### **Conclusion**

> Scaling transforms the dataset from numerically imbalanced to model-ready, enabling correct and efficient machine learning.


### **Impact of scaling on ML algorithms:**

#### **Impact of Scaling on ML Algorithms**

- Scaling puts all numerical features on the **same range**
- Prevents large-value features from **dominating the model**

##### **Algorithms that NEED Scaling**
- **Logistic Regression** → faster and stable convergence
- **KNN** → fair distance calculation
- **SVM** → correct margin and boundary
- **PCA** → meaningful principal components
- **Neural Networks** → stable and faster training

##### **Algorithms that DO NOT Need Scaling**
- **Decision Trees**
- **Random Forest**
- **Gradient Boosting**

##### **Key Takeaway:**
> Scaling is essential for distance-based and gradient-based algorithms, but not required for tree-based models.
