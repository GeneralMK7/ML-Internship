### **Plotting histograms for Numerical Data:**

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
```

```python
release_year_data = df['release_year']
sns.histplot(data=release_year_data,bins=30,alpha=0.7,edgecolor='black',color='teal')
plt.xlabel('Release Year',size=14)
plt.ylabel('Number of Movies',size=14)
plt.title('Distribution of Movies by Release Year',size=18)
plt.savefig('distribution_of_movies_by_release_year.png')
plt.show()
```

### **Outliers Concept using Box Plot:**

```python
sns.boxplot(x=df['release_year'])
plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.savefig('Distribution_of_release_years_by_count.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='type', y='release_year')

plt.title("Release Year Distribution by Type")
plt.xlabel("Type")
plt.ylabel("Release Year")
plt.savefig('Distribution_of_release_years_by_type.png')
plt.show()
```

### **Correlation Heatmap to understand feature Relationships:**

```python
cross_tab = pd.crosstab(df['type'], df['rating'])

plt.figure(figsize=(12,6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='Blues')

plt.title("Heatmap of Type vs Rating")
plt.xlabel("Rating")
plt.ylabel("Type")
plt.savefig('Type_vs_rating_heatmap.png')
plt.show()
```

```python
df_exploded = df.assign(
    country=df['country'].str.split(', ')
).explode('country')
```

```python
top10 = df_exploded['country'].value_counts().head(20).index
filtered_df = df_exploded[df_exploded['country'].isin(top10)]
```

```python
cross_tab = pd.crosstab(filtered_df['type'], filtered_df['country'])
plt.figure(figsize=(12,6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='Greens')

plt.title("Type vs Country (Top 20)")
plt.xlabel("Country")
plt.ylabel("Type")
plt.savefig('Type_vs_country_heatmap.png')
plt.show()
```

## **Insights for each Visualization:**

### **1. Movies by Rating**
Netflix's shows and movies skew super matureTV-MA and TV-14 take up most of the space, while kid stuff like G or TV-Y is barely there. It's clear that they're chasing teens and adults, not families.

### **2. Movies by Release Year (Histogram)**
Titles explode after 2010, with almost nothing pre-1980. Netflix's lineup is all about fresh, modern content, not dusty old classics.Like they have more films which came in like around 2016.

### **3. Movies by Type**
Movies are more than TV shows in numbers  While TV Shows are important, the platform remains more movie-centric.

### **4. Release Years (Box Plot)**
The median release year is very recent, and the distribution is heavily left-skewed, with many old titles acting as outliers. Most of the content lies within a narrow modern time window.

### **5. Release Years by Type (Movie vs TV Show)**
TV shows edge out movies on recency, but movies have more ancient outliers.

### **6. Top 10 Countries**
US owns many Movies and TV Shows, and then comes India and UK. Netflix is super US-heavy with limited global spread.

### **7. Top 10 Genres**
Genres like International Movies, Dramas, and Comedies dominate Netflixs catalog. Niche genres appear much less frequently.
### **8. Heatmap: Type vs Country**
Across nearly all top countries, Movies outnumber TV Shows. The United States leads in both categories, followed by India and the UK.

### **9. Heatmap: Type vs Rating**
Movies are mostly concentrated in TV-MA, R, and PG-13, while TV Shows dominate TV-MA and TV-14. Family-friendly ratings appear rarely in both categories.

### **All-over Summary**:
Netflixs content strategy is heavily skewed toward modern, mature, and globally appealing content, with a strong dominance of movies over TV shows. The platform prioritizes adult audiences, recent releases, and content from a small number of leading production countries, particularly the United States.
