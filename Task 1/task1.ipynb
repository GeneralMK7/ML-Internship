## 5. **Identifying Target Variable and Input Features**

- The target variable is the column that the model aims to predict.
- In the Titanic dataset, **Survived** is the target variable.
- It represents a binary classification problem (0 = Did not survive, 1 = Survived).
- Input features are the columns used to make predictions.
- Features such as **Age, Sex, Pclass, Fare, SibSp, Parch, and Embarked** are suitable for modeling.
- Identifier-like columns such as **PassengerId** and **Ticket** are not useful for prediction.


```python
#If we want to predict the survival based on the dataset we can deop 'PassengerId' and 'ticket'
train_dataset_filter = train_dataset.drop(columns=['PassengerId', 'Ticket']) # Once run,we cannot take back the data if put into same variable
```

```python
input_features = train_dataset_filter.drop(columns='Survived')
input_features
```
